```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import make_classification
```

```python
movies = pd.read_csv('n_movies.csv')
```

```python
movies
```

```python
movies = movies.drop_duplicates(subset = "title", keep="first")
```

```python
movies
```

```python
movies.dropna()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
features = ['genre','rating', 'votes']
def combine_features(row):
    rate = str(row['rating'])
    vote = str(row['votes'])
    genre = str(row['genre'])
    return(genre +" "+ rate + " " + vote)

movies["combined_features"] = movies.apply(combine_features,axis=1)
cv = CountVectorizer()
matrix = cv.fit_transform(movies["combined_features"])
cosine_sim = cosine_similarity(matrix)

def get_title(index):
    return movies[movies.index == index]["title"].values[0]

def get_index(title):
    return movies[movies.title == title].index.values[0]

user_movie = "Cobra Kai"
movie_index = get_index(user_movie)
similar_movies = list(enumerate(cosine_sim[movie_index]))

similar_movies = sorted(similar_movies,key=lambda x:x[1],reverse=True)[1:]

i=0
print("Top 5 similar movies to "+user_movie+" are:\n")
for element in similar_movies:
    print(get_title(element[0]))
    i=i+1
    if i>=5:
        break
```
